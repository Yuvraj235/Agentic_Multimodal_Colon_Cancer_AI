```python
import os
import cv2
from tqdm import tqdm

PROCESSED_ROOT = "../data/processed/hyper_kvasir_clean"

bad_images = []

for root, _, files in os.walk(PROCESSED_ROOT):
    for f in tqdm(files):
        if f.lower().endswith((".jpg", ".jpeg", ".png")):
            path = os.path.join(root, f)
            img = cv2.imread(path)
            if img is None or img.size == 0:
                bad_images.append(path)

print("Corrupted images found:", len(bad_images))
```

```python
for path in bad_images:
    os.remove(path)

print("Removed corrupted images.")
```

```python
valid_images = set()

for root, _, files in os.walk(PROCESSED_ROOT):
    for f in files:
        if f.lower().endswith((".jpg", ".jpeg", ".png")):
            valid_images.add(f)

print("Total valid processed images:", len(valid_images))
```

```python
import os
import pandas as pd

LABELS_PATH = "../data/raw/hyper-kvasir-labeled-image/image-labels.csv"
labels_df = pd.read_csv(LABELS_PATH)

labels_df.head()
```

```python
PROCESSED_ROOT = "../data/processed/hyper_kvasir_clean"

valid_images = set()

for root, _, files in os.walk(PROCESSED_ROOT):
    for f in files:
        if f.lower().endswith((".jpg", ".jpeg", ".png")):
            valid_images.add(f)

print("Total valid processed images:", len(valid_images))
```

```python
labels_df["image_name"] = labels_df["Video file"].astype(str) + ".jpg"

clean_df = labels_df[labels_df["image_name"].isin(valid_images)].copy()

print("Before cleaning:", len(labels_df))
print("After cleaning:", len(clean_df))
```

```python
CLEAN_LABELS_PATH = "../data/processed/clean_labels.csv"

clean_df.drop(columns=["image_name"], inplace=True)
clean_df.to_csv(CLEAN_LABELS_PATH, index=False)

print("✅ Clean labels saved to:", CLEAN_LABELS_PATH)
```
